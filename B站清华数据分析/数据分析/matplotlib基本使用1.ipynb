```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.arange(10)
y1 = np.random.randint(0,10,10)
x1,y1
```

```python
plt.plot(y1)
```

```python
plt.plot(x1,y1)
```

```python
data1 = pd.DataFrame({"a":x1,"b":y1,"c":np.random.permutation(10)})
data1
```

```python
plt.plot("a","b","",data=data1)
```

```python
plt.plot("a","c","",data=data1)
```

```python
lines = plt.plot(x1,y1,x1,np.random.permutation(10))
line0 = lines[0]
line0.set_color("r")
line0.set_linewidth(4)
line0.set_alpha(0.7)
line0.set_linestyle("--")
```

```python
y1 = [np.random.randint(0,10) for x in range(10)]
y2 = [np.random.randint(0,10) for x in range(10)]
lines = plt.plot(range(10),y1,range(10),y2)
plt.setp(lines,lw=4)
```
