```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager
font = font_manager.FontProperties(fname=r"C:\\Windows\\Fonts\\msyh.ttc",size=10)
```

```python
oses = {
    'windows7':60.86,
    'windows10': 18.46,
    'windows8': 3.61,
    'windows xp': 10.3,
    'mac os': 6.78,
    '其他': 1.12
}
```

```python
patches,texts,autotexts = plt.pie(oses.values(),autopct="%.2f%%",labels=oses.keys(),textprops={"fontproperties":font,"size":8},shadow=True,explode=(0,0.1,0,0,0,0))
for text,autotext in zip(texts,autotexts):
    autotext.set_color("w")
#     autotext.set_size(8)
```

# 笔记：
饼图是一个划分为几个扇形的圆形统计图表，用于描述量、频率或百分比之间的相对关系的。
在`matplotlib`中，可以通过`plt.pie`来实现，其中的参数如下：
1. `x`：饼图的比例序列。
2. `labels`：饼图上每个分块的名称文字。
3. `explode`：设置某几个分块是否要分离饼图。
4. `autopct`：设置比例文字的展示方式。比如保留几个小数等。
5. `shadow`：是否显示阴影。
6. `textprops`：文本的属性（颜色，大小等）。
7. 其他参数：https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie

**返回值：**
1. `patches`：饼图上每个分块的对象。
2. `texts`：分块的名字文本对象。
3. `autotexts`：分块的比例文字对象。
