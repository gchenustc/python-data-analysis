```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.font_manager import FontProperties
font = FontProperties(fname="C:\Windows\Fonts\STSONG.TTF",size=20)
```

# 普通条形图

```python
movies = {
    "流浪地球":40.78,
    "飞驰人生":15.77,
    "疯狂的外星人":20.83,
    "新喜剧之王":6.10,
    "廉政风云":1.10,
    "神探蒲松龄":1.49,
    "小猪佩奇过大年":1.22,
    "熊出没·原始时代":6.71
}
```

```python
df = pd.Series(movies)
df
```

```python
plt.figure(figsize=(10,4))
plt.bar(df.index, df.values, width=0.6)
plt.xticks(fontproperties=font,size=8)
font.set_size(8)
plt.yticks(np.arange(0,45,5),["第 %d 亿" % x for x in np.arange(0,45,5)], fontproperties=font)
```

# 横向条形图

```python
df1 = pd.DataFrame({"names": df.index, "values": df.values})
df1
```

```python
plt.figure(figsize=(4,5),dpi=200,facecolor="y",linewidth=3,edgecolor="b")
plt.barh("names","values",data=df1,height=0.5)
plt.yticks(fontproperties=font)
plt.xticks(range(0,45,5),["%d亿" %x for x in range(0,45,5)],fontproperties=font)
```

# 堆叠条形图

```python
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
groupNames = ('G1','G2','G3','G4','G5')
plt.bar(groupNames, menMeans, label="男性得分")
plt.bar(groupNames, womenMeans, bottom=menMeans, label="女性得分")
plt.legend(prop=font)
```

# 分组条形图

```python
movies = {
    "流浪地球":[2.01,4.59,7.99,11.83,16],
    "飞驰人生":[3.19,5.08,6.73,8.10,9.35],
    "疯狂的外星人":[4.07,6.92,9.30,11.29,13.03],
    "新喜剧之王":[2.72,3.79,4.45,4.83,5.11],
    "廉政风云":[0.56,0.74,0.83,0.88,0.92],
    "神探蒲松龄":[0.66,0.95,1.10,1.17,1.23],
    "小猪佩奇过大年":[0.58,0.81,0.94,1.01,1.07],
    "熊出没·原始时代":[1.13,1.96,2.73,3.42,4.05]
}
df = pd.DataFrame(movies)
df
```

```python
plt.figure(figsize=(10,5), dpi=250, facecolor="y", edgecolor="b", linewidth=3)
column_width = 0.15
for index in df.index:
    plt.bar(np.arange(len(df.columns))-(index-2)*column_width, df.loc[index], width=column_width, label="第%d天票房" % (index+1))
plt.legend(prop=font)
plt.xticks(np.arange(len(df.columns)), df.columns, fontproperties=font)
plt.show()
```

# 笔记：
1. 绘制条形图使用的是plt.bar方法。这个方法有很多参数：
    * x：在x轴上的坐标，可以为一个列表/数组/字符串。
    * y：在y轴上的坐标，可以为一个列表/数组/字符串。
    * data：如果传递了data，那么x和y就可以为data中的key。比如data是一个DataFrame对象，那么x和y就是这个DataFrame对象的某个列的名字。
2. 官方文档：https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar
3. 横向条形图：plt.barh。然后y代表的是之前bar的x，width代表的是bar的height。其他参数都差不多。
4. 分组条形图：matplotlib并没有提供专门的分组条形图的API来实现。而是需要设置这个坐标来实现。绘制的规则是，一次性绘制所有电影第1天的数据，第二天的数据2...，然后要找到规律，把这个条形图的位置算好。
5. 堆叠条形图：matplotlib并没有提供专门的堆叠条形图的API来实现，我们在绘制往上堆叠的数据的时候，可以使用bottom参数，来往上偏移指定的距离。这样就可以实现堆叠的效果了。
