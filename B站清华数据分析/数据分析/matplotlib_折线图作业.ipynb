# 折线图作业要求：
以下是长沙某一个月的天气数据，按照时间的顺序绘制成折线图，其中数据highest是最高温度，lowest是最低温度。最高温度线条用红色，最低温度线条用蓝色。
具体的坐标点，用圆点marker表示。
把x轴的时间刻度按照1-31标记出来，并且标记x轴和y轴的标题。
图的标题是“长沙5月份气温走势”。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties 
```

```python
highest = [26,21,26,26,22,20,17,19,22,28,30,28,24,28,25,26,25,26,25,23,24,30,32,31,30,27,26,27,29,25,25]
lowest =  [17,13,17,18,18,17,14,15,16,18,19,20,18,18,20,20,20,20,20,16,17,19,21,24,24,23,20,18,19,18,19]
days = np.arange(1,len(highest)+1)
data = pd.DataFrame({"highest":highest,"lowest":lowest,"days":days})
font=FontProperties(fname="C:\Windows\Fonts\simkai.ttf",size=10)
```

```python
fig = plt.figure(dpi=200,figsize=(15,5))
plt.grid(True)
plt.title("长沙五月份温度走势",fontproperties=font,fontsize=20)
plt.plot("days","lowest","",data=data,marker="o",label="最低气温")
plt.plot("days","highest","",data=data,marker="o",color="r",label="最高气温")
plt.legend(prop=font,loc="upper center",ncol=2,frameon=False)
# x
plt.xticks(days,days,fontsize=12)
plt.xlim(1,31)
plt.xlabel("日期(天)",fontproperties=font,fontsize=20)
# y
plt.ylim(10,35)
plt.ylabel("温度(°C)",fontproperties=font,fontsize=20)
#注释
for x,y1,y2 in zip(data.iloc[:,2],data["highest"],data["lowest"]):
    plt.annotate(text=y1,xy=(x,y1),xytext=(x,y1+0.1),fontsize=12)
    plt.annotate(text=y2,xy=(x,y2),xytext=(x,y2+0.1),fontsize=12)
```
