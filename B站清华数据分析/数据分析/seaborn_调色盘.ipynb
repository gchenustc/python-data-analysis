```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#查看调色盘样式_定性调色盘
sns.choose_colorbrewer_palette("qualitative")
```

```python
#查看默认调色盘
plaette = sns.color_palette()
sns.palplot(plaette)
```

```python
#默认调色盘的几种样式
plaette = sns.color_palette("pastel")
sns.palplot(plaette)
```

```python
#hls
palette = sns.color_palette("hls",n_colors = 20)
sns.palplot(palette)
palette = sns.color_palette("hls",50)
sns.palplot(palette)
```

```python
sns.palplot(sns.hls_palette(20,h=0.2,l=0.6,s=0.4))
sns.palplot(sns.hls_palette(20,h=0.2,l=0.6,s=0.8))
sns.palplot(sns.hls_palette(20,h=0.2,l=0.3,s=0.8))
```

```python
#分类调色盘
sns.palplot(sns.color_palette("Paired"))
```

```python
#连续的调色盘
sns.palplot(sns.color_palette("Blues",15))
sns.palplot(sns.color_palette("Blues_r",15))
```

```python
sns.choose_colorbrewer_palette("sequential")
```

```python
#离散的调色盘
sns.choose_colorbrewer_palette("diverging")
```

```python
data = pd.read_csv("dataset/tips.csv")
sns.set(palette="pastel")
sns.barplot(x = "day", y = "tip", data=data, estimator=sum, order=["Thur","Fri","Sat","Sun"])
```

```python
with sns.color_palette("Reds"):
    sns.barplot(x = "day", y = "tip", data=data, estimator=sum, order=["Fri","Thur","Sun","Sat"])
```
