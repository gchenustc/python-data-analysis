```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import *
import os
```

```python
x = np.linspace(0,20)
plt.plot(x, np.sin(x), x, np.cos(x))
```

```python
x = np.linspace(0,20)
plt.plot(np.sin(x),label="sin")
plt.plot(np.cos(x),label="cons")
plt.plot(np.random.permutation(50)/50,label="permutation")
plt.legend()
```

```python
plt.figure(figsize=(10,5),dpi=200)
plt.subplot(221)
plt.plot(np.sin(x),c="r")
plt.plot(np.sin(x-pi/4))
plt.subplot(222)
plt.plot(np.cos(x),c="k")
plt.subplot(223)
plt.plot(np.tan(x),c="g")
plt.subplot(224)
plt.plot(x)
plt.plot([pow(i,2) for i in x])
```

```python
x = np.linspace(0,20)
plt.figure(figsize=(10,5),dpi=200)
plt.style.use("seaborn-pastel")
fig,axis = plt.subplots(2,2)
ax1 = axis[0,0]
ax2 = axis[0,1]
ax3 = axis[1,0]
ax4 = axis[1,1]
ax1.plot(x)
ax2.plot(np.sin(x))
ax3.plot(np.cos(x))
ax4.plot(np.sin(2*x))
```
