```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# 读取数据
data = pd.read_csv("dataset/fmri.csv")
data.tail()
```

```python
# 绘制散点图
sns.relplot(x="timepoint", y="signal", data=data)
```

```python
# 绘制散点图,ci=False(默认为True)画不带置信区间的折线图
sns.relplot(kind="line", x="timepoint", y="signal", hue="event", hue_order=["stim","cue"],style="event",col="region",data=data)
```

# 笔记
1.sns.relplot(kine="line")代表画折线图

2.seaborn的折线图不需要x与y一一对应，一个x可以有多个y，最后展现的折现代表y轴上的平均值

3.ci=False代表关闭置信区间的阴影
