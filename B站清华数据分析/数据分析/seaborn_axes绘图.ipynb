```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("dataset/tips.csv")
data.head(2)
```

```python
axes = sns.scatterplot(x="total_bill",y="tip",data=data)
type(axes)
#可以在matplotlib官网搜索axes相关参数调节图片
```

```python
# FacetGrid绘图中的x和y参数不能传入pandas.Series
axes1 = sns.relplot(x="total_bill",y="tip",data=data)
type(axes1)
```

```python
fig,[ax1,ax2] = plt.subplots(1,2,figsize=(25,7))
sns.scatterplot(x="total_bill",y="tip",data=data,ax=ax1)
sns.barplot(x="day",y="total_bill",data=data,ax=ax2)
```

# 笔记：
1. FacetGrid，可以通过col和row等参数来一次性构建多个图形。
2. relplot/catplot/lmplot等函数底层都是使用FacetGrid技术。
3. 凡是函数名中已经明确了这个图的类型，都是直接使用Axes进行绘图。
4. FacetGrid绘图的x和y参数，必须为DataFrame的列的名字。而直接使用Axes绘图，参数更加灵活。
5. 直接使用Axes绘图，可以非常方便的跟matplotlib结合使用。
