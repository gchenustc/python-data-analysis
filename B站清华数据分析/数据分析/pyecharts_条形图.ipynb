```python
from pyecharts.charts import Bar
import pyecharts.options as opts
from faker import Faker
```

```python
# 反转xy轴
c = (Bar()
     .add_xaxis(Faker.choose())
     .add_yaxis("商家A",Faker.values())
     .add_yaxis("商家B",Faker.values())
     .reversal_axis()
)
c.render_notebook()
```

```python
# 堆积条状图
c = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("商家A",Faker.values(),stack=True)
    .add_yaxis("商家B",Faker.values(),stack=True)
    .reversal_axis()
)
c.render_notebook()
```

```python
# 单系列柱间距离调整 category_gap
# 同系列柱间距离调整 gap
c = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("商品A",Faker.values(),category_gap="20%",gap=0.1)
    .add_yaxis("商品B",Faker.values(),category_gap="20%",gap=0.1)
)
c.render_notebook()
```
