### 用pandas从scores.csv读取出来，形成一个DataFrame对象。
绘制化学成绩的直方图（chem列）。
标记x轴的坐标。
标记每个条形上的具体数值。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname="C:\Windows\Fonts\simkai.ttf",size=10)
```

```python
# 以三班化学成绩为例
# 表格处理
data = pd.read_csv("scores.csv")
class3 = data[data["class"]== 3]
class3_chem = class3["chem"]
class3_chem.index = np.arange(len(class3_chem))
class3_chem.head()
```

```python
# 创建figure
fig = plt.figure(dpi=300, figsize=(15,5), facecolor="#a1b2c3d4")
# 绘图
ax = plt.subplot(111)
nums,bins,patchs = ax.hist(class3_chem,bins=np.arange(5,50,3),edgecolor="k")
#标题和图例
plt.title("3班化学成绩分布",fontproperties=font,fontsize=20)
# x and ylabel
ax.set_xlabel("分数",fontproperties=font,fontsize=12)
ax.set_ylabel("人数",fontproperties=font,fontsize=12)
# 细节
ax.set_xticks(np.arange(5,50,3))
ax.set_xlim(7,48)
ax.minorticks_on()
ax.set_ylim(0,15)
    #标记每个柱子上的具体数值
for x_,y_ in zip(np.arange(5,50,3),nums):
    ax.annotate(y_,(x_,y_),(x_+1,y_),fontsize=15)
```
