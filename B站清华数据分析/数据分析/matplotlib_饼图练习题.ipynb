```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname="C:\Windows\Fonts\simkai.ttf",size=10)
browsers = {
    "Chrome": 0.5098,
    "Internet Explorer": 0.1218,
    "FireFox": 0.1147,
    "Edge": 0.0415,
    "Safari": 0.0372,
    "其他浏览器": 0.075
}
```

```python
#绘制gigure
figure = plt.figure(dpi=300,figsize=(10,10),facecolor="y",edgecolor="k",linewidth=2,frameon=True)
#绘制axes
ax = figure.add_axes([0.9,0.9,0.8,0.8])
patchs,texts,autotexts = ax.pie(browsers.values(), autopct="%.2f%%", labels=browsers.keys(), shadow=True, normalize=False, explode=(0.05,0,0,0,0,0), textprops={"fontproperties":font,"fontsize":10})
#图例和标题
ax.legend(prop=font, fontsize=12, loc="upper right", frameon=False)
plt.title("各大浏览器使用人数比例",fontproperties=font,fontsize=20)
#细节
for text in texts[3:5]:
    text.set_size= (12)
    text.set_color("b")
for autotext in autotexts:
    autotext.set_color("w")
```

### 
normalize：是否归一化。布尔值或None。默认值为None。
True：完全饼图，对x进行归一化，sum(x)==1。
False：如果sum(x)<=1，绘制不完全饼图。如果sum(x)>1，抛出ValueError异常。
None：如果sum(x)>=1，默认值为True。如果sum(x)<1，默认值为False。
注意：未来版本（当前版本3.3.3），默认值将改为True。绘制不完全饼图，需要明确传递normalize=False。
