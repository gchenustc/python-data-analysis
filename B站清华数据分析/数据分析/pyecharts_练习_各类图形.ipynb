```python
# Timeline()时间轴
tl = Timeline()

day_begin=datetime.date(2021,11,1)
day_end=datetime.date(2021,11,5)
cate=Faker.choose()

for day in range((day_end-day_begin).days+1):
    date = day_begin + datetime.timedelta(day)
    bar=(
        Bar()
        .add_xaxis(cate)
        .add_yaxis("",Faker.values())
        )
    tl.add(bar,date)
tl.render_notebook()
```

```python
# grid布局
x_data = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
y_data = [123, 153, 89, 107, 98, 23]


bar = (Bar()
       .add_xaxis(x_data)
       .add_yaxis('', y_data)
       )

line = (Line()
        .add_xaxis(x_data)
        .add_yaxis('', y_data)
        )
grid=(
    Grid()
    .add(bar,grid_opts=opts.GridOpts(pos_bottom="70%",pos_left="50%"))
    .add(line,grid_opts=opts.GridOpts())

)
grid.render_notebook()
```

```python
# 主题
theme_list = ['chalk',
              'dark',
              'essos',
              'infographic',
              'light',
              'macarons',
              'purple-passion',
              'roma',
              'romantic',
              'shine',
              'vintage',
              'walden',
              'westeros',
              'white',
              'wonderland']
x = Faker.choose()
y = Faker.values()
page = Page()
for theme in theme_list:
    bar = (
            Bar(init_opts=opts.InitOpts(theme=theme,height="180px",width="300px"))
            .add_xaxis(x)
            .add_yaxis("",y)
            .set_global_opts(title_opts=opts.TitleOpts("Theme-{}".format(theme),title_textstyle_opts=opts.TextStyleOpts(font_size=10)))
    )
    page.add(bar)
page.render_notebook()
```

```python
# 背景颜色
# 虚假数据
x_data = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
y_data_1 = [123, 153, 89, 107, 98, 23]
y_data_2 = [231, 321, 135, 341, 245, 167]

# 背景颜色配置
bar = (
    Bar(init_opts=opts.InitOpts(bg_color='rgba(100,200,200,200)',theme="dark"))
    .add_xaxis(x_data)
    .add_yaxis('', y_data_1)
    .add_yaxis('', y_data_2)
)

bar.render_notebook()
```
