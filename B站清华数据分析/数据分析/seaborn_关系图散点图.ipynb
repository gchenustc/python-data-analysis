```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
tips = pd.read_csv("dataset/tips.csv")
tips.head()
```

## 绘图

```python
sns.relplot(kind="scatter",x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(kind="scatter",x="total_bill",y="tip",hue="day",size="size",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="day",hue_order=["Sat","Sun","Fri","Thur"],data=tips)
```

```python
sns.relplot(kind="scatter",x="total_bill",y="tip",hue="day",col="time",row="sex",data=tips)
```

```python
sns.relplot(kind="scatter",x="total_bill",y="tip",col="day",col_wrap=2,data=tips)
```

# 笔记

1.散点图绘制：sns.relplot(),默认情况下不加Kind属性就是散点图，也可以sns.relplot(kind="scatter")

2.属性：
    * x,y,data
    
    * col,row :指定列或者行按照某个属性分类
    
    * col_wrap：指定分成多少列
    
    * hue：在一个散点图中用点的颜色分类并且生成图例
    
    * hue_order：hue_order=[]代表hue的顺序
    
    * size：在一个散点图中用点的大小分类并且生成图例
