```python
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("dataset/tips.csv")
g = sns.lmplot(x="total_bill",y="tip",data=data,col="day",col_wrap=2)
type(g)
```

```python
g = sns.regplot( x=data["total_bill"],y=data["tip"] )
type(g)
```

# 笔记:
1. regplot和lmplot都可以绘制线性回归图。
2. regplot的x和y可以为numpy的数组或者是Series对象。但是lmplot不可以
3. 一些其他的参数，比如order，robust等，可以调节线性的准确度。
4. regplot的类型是axes，lmplot的类型是facetgird
