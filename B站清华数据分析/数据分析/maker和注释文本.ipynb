```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
x = np.linspace(0,20,50)
y = np.sin(x)
plt.plot(x,y,marker="o",markerfacecolor ="r", markeredgecolor="b", markersize=10)
plt.figure(figsize=(12,5))
```

```python
y = np.sin(np.arange(20))
plt.plot(y,marker="o",markersize="3")
plt.annotate("0,0",xy=(0,0),xytext=(-0.5,-0.1))
plt.annotate("0,0",xy=(0,0),xytext=(-0.5,-0.6),arrowprops={"width":3, "headwidth":6, "headlength": 4, "shrink": 0.3, "facecolor": "r", "edgecolor": "y"})
```

```python
y = pd.Series(np.sin(np.arange(1,21)), index=range(1,21))
y
```

```python
plt.plot(y,marker="o",markersize="3")
font = FontProperties(fname=r"C:\\Windows\\Fonts\\msyh.ttc",variant="small-caps", style='italic')
font.set_size(9)
for index,value in dict(y).items(): 
    plt.annotate("(%d,%.2f)" % (index,value), xy=(index, value), xytext=(index-0.5, value-0.1), fontproperties=font)
```
