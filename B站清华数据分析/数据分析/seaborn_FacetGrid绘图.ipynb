```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# rc={}是调用matplotlib配置参数，详情参考matplotlib
sns.set(rc={"font.sans-serif":"SimHei"})
data = pd.read_csv("dataset/tips.csv")
g1 = sns.FacetGrid(data,col="day",col_wrap=2,hue="sex",height=4,aspect=1.5)
g1.map(sns.regplot,"total_bill","tip")
# 设置图例
g1.add_legend(title="性别")
for English,Chinese in zip(g1._legend.texts,["女性","男性"]):
    English.set_text(Chinese)
# 设置标题
g1.set_titles(template="日期:{col_name}")
# 设置轴标题\
g1.set_axis_labels("总消费","小费")
# 设置轴刻度
g1.set(xticks=np.arange(0,60,5),yticks=np.linspace(0,10,11))
# 设置轴
g1.set_xticklabels(["%d$" % x for x in np.arange(0,60,5)])
g1.set_yticklabels(["%d$" % x for x in np.linspace(0,10,11)])
# set
g1.set(facecolor="y")
# fig
g1.fig.set_dpi(100)
```
