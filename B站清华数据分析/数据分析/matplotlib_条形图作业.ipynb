```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties 
```

```python
data = {
    "普通本科":[721,738,749,761,791],
    "中等职业教育": [620,601,593,582,557],
    "普通高中": [797,797,803,800,793]
}
df = pd.DataFrame(data).T
font=FontProperties(fname="C:\Windows\Fonts\simkai.ttf",size=10)
df
```

```python
# 绘图
fig = plt.figure(figsize=(15,5),dpi=300)
plt.grid(True)
plt.title("2014-2018普通本科，中等职业教育，普通高中招生人数", fontproperties=font, fontsize=20)
for i in range(len(df)):
    ax = fig.add_subplot(111)
    ax.bar(np.arange(5)+0.2*(i-1), df.iloc[i], width=0.2, label=df.index[i])
    ax.set_ylim(0,1000)
# 图例   
plt.legend(frameon=False, loc="upper right",ncol=3, fontsize=3,prop=font)
# 细节
    # x axis
plt.xticks(np.arange(5),np.arange(2014,2019),fontsize=12)
plt.xlabel("年份",fontproperties=font,fontsize=15)
        # x轴label位置
ax.xaxis.set_label_coords(0.97,-0.020)
    # y axis
        # y轴单位位置
# plt.text(-0.73,1050,"万人",fontdict={"fontproperties":font,"size":15})
plt.show()
```
