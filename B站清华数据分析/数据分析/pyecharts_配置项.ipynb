```python
from pyecharts.charts import Bar,Scatter
import pyecharts.options as opts
from faker import Faker
from pyecharts.commons.utils import JsCode
```

```python
bar = (Bar()
    .add_xaxis(["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"])
    .add_yaxis("商家A", [5, 20, 36, 10, 75, 90])
      )
bar.render_notebook()
```

```python
# init_opts
c = (
    Bar(
        init_opts=opts.InitOpts(
            width="400px",
            height="250px",
            bg_color="#bbb",
            animation_opts=opts.AnimationOpts(animation=False)
        ))
     .add_xaxis(Faker.choose())
     .add_yaxis("商品",Faker.values())
    )
c.render_notebook()
```

```python
# 标题
c=(
    Bar(
        init_opts=opts.InitOpts(
            width="400px",
            height="250px",
            bg_color="#aaa"
    ))
    .add_xaxis(Faker.choose())
    .add_yaxis("商品",Faker.values())
    .set_global_opts(
    title_opts=opts.TitleOpts(
        title="商品售出情况",
        subtitle="子标题",
        pos_left="20px",
        title_textstyle_opts=opts.TextStyleOpts(font_size=10)
    ))
)
c.render_notebook()
```

```python
# 图例
c=(
    Bar(
        init_opts=opts.InitOpts(
        width="500px",
        height="300px"
    ))
    .add_xaxis(Faker.choose())
    .add_yaxis("第一季度",Faker.values(),gap=0)
    .add_yaxis("第二季度",Faker.values(),gap=0)
    .set_global_opts(
        legend_opts=opts.LegendOpts(
            selected_mode="single",
            orient="vertical",
            pos_right="30px"
    ))
)
c.render_notebook()
```

```python
# visualmap_opts
values=list(zip(Faker.values(),Faker.values()))
c = (
    Scatter(
        init_opts=opts.InitOpts(
            width="450px",
            height="275px",
            bg_color="#1bc2de3f"
    ))
    .add_xaxis(Faker.choose())
    .add_yaxis("品类1",values)
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            type_="color",
            range_text=["大","小"]
        ),
        legend_opts=opts.LegendOpts(
            pos_right="20px"
    ))
)
    
c.render_notebook()
```

```python
# 轴
c = (
    Bar(
        init_opts=opts.InitOpts(width="500px",height="300px")
    )
    .add_xaxis(
        [
        "名字很长的X轴标签1",
        "名字很长的X轴标签2",
        "名字很长的X轴标签3",
        "名字很长的X轴标签4",
        "名字很长的X轴标签5",
        "名字很长的X轴标签6",
        ]
    )
    .add_yaxis("商家A", [10, 20, 30, 40, 50, 40],category_gap="50%",gap=0,)
    .add_yaxis("商家B", [20, 10, 40, 30, 40, 50],category_gap="50%",gap=0)
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            name="商家名称",
            axislabel_opts=opts.LabelOpts(rotate=-35),
            axisline_opts = opts.AxisLineOpts(symbol="arrow",linestyle_opts=opts.LineStyleOpts(width=3)),
            axistick_opts = opts.AxisTickOpts(is_inside=True,length=3),
            axispointer_opts = opts.AxisPointerOpts(is_show=True,type_="line"),
            splitline_opts = opts.SplitLineOpts(is_show=True)
        ),
        yaxis_opts=opts.AxisOpts(
            axispointer_opts = opts.AxisPointerOpts(is_show=True,type_="line"),
            axisline_opts = opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(width=2)),
            axistick_opts = opts.AxisTickOpts(is_inside=True,length=3),
            splitline_opts = opts.SplitLineOpts(is_show=True)
        ),
        legend_opts = opts.LegendOpts(pos_right="10px",orient="vertical")
    )
)
c.render_notebook()
```
