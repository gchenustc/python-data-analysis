```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import *
import os
```

```python
x = np.linspace(0,4*pi,32)
y = np.sin(x)
data1 = pd.DataFrame({"x":x, "y":y})
plt.figure(figsize=(10,5),dpi=200)
plt.plot("x","y","", data = data1, marker="o")
plt.savefig("D:"+os.path.sep+"sin.png")
```
